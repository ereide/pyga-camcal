```python
import sympy

from sympy import symbols, sin, cos, sinh, cosh, init_printing

init_printing()
```

```python
from sympy import I
```

```python
theta, phi, omega, N, P, t_par_n, t_nor_n, k= symbols("theta phi omega N P t_par t_nor k")

```

```python
I3n = symbols("I3n")
```

```python
B = theta*P + (t_par_n + t_nor_n) + omega * N 
```

```python
(B * B* B).expand().subs(N*N, 1).subs(P*P, -1).subs(P*t_nor_n, I3n)
```

```python




R_expected = ((cos(phi) + sin(phi) * P)*(cosh(omega) + sinh(omega) * N + sinh(omega)/omega * t_nor_n)    +  
                    1/k*
                    ( (-omega * sin(phi) * cosh(omega) + phi * cos(phi) * sinh(omega)) * P 
                    + ( omega * cos(phi) * sinh(omega) + phi * sin(phi) * cosh(omega)) * t_par_n))

R_expected_tilde = ((cosh(omega) - sinh(omega) * N - sinh(omega)/omega * t_nor_n)*(cos(phi) - sin(phi) * P) -
                    1/k*
                    ( (-omega * sin(phi) * cosh(omega) + phi * cos(phi) * sinh(omega)) * P 
                    + ( omega * cos(phi) * sinh(omega) + phi * sin(phi) * cosh(omega)) * t_par_n))


```

```python
expr = (R_expected*R_expected_tilde).expand().trigsimp().simplify()

```

```python
expr.trigsimp()
```

```python
expr.subs(N*N, 1).subs(P*P, -1).subs(P*t_nor_n, I3n)
```
