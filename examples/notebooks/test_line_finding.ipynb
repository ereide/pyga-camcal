```python
from pygacal.rotation import minimizeError
from pygacal.rotation.mapping import BivectorLineImageMapping

from clifford_tools.common.g3c.core import random_line, generate_translation_rotor
from clifford_tools.common.g3.core import generate_rotation_rotor

import numpy as np
from clifford.g3c import *
```

```python
line_list_a = [random_line() for i in range(100)]
r = (generate_rotation_rotor(np.pi/4,e1+e2,e1)).normal()
line_list_b = [(r*l*~r).normal() for l in line_list_a]

R_min, Nint = minimizeError((line_list_a, line_list_b), 
                                BivectorLineImageMapping, 
                                x0 = None)
line_list_mapped = [(R_min*l*~R_min).normal() for l in line_list_a]
```

```python
for a,b in zip(line_list_mapped,line_list_b):
    print(a-b)
```
